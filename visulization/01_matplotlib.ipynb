### Line plot Basic

```python
import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(1,11)
y = 2*x
z = 3*x
a = np.exp(x)
```

```python
plt.figure(figsize = (8,6))
plt.plot(x,y, color = 'red', linestyle='--', linewidth = 2, marker = '*', markersize = 10, markerfacecolor='black', label = 'Area mapping')
plt.plot(x,z, color = 'green', linestyle='dotted', linewidth = 2, marker = 'o', markersize = 10, markerfacecolor='yellow', label = 'Volums mapping')
plt.plot(x,a, color = 'black', linestyle=':', linewidth = 2, marker = 'D', markersize = 10, markerfacecolor='blue', label = 'exp mapping')
plt.title('line plot', fontweight = 'bold', color = 'red')
plt.xlabel('X - label')
plt.ylabel('Y - label')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,1))
plt.grid(True, linestyle = '--', alpha = 0.5, color = 'red')
plt.show()
```

```python
plt.figure(figsize = (15,5))
plt.subplot(1,3,1)
plt.plot(x,y, color = 'red', linestyle='--', linewidth = 2, marker = '*', markersize = 10, markerfacecolor='black', label = 'Area mapping')
plt.title('line plot', fontweight = 'bold', color = 'red')
plt.xlabel('X - label')
plt.ylabel('Y - label')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,1))
plt.grid(True, linestyle = '--', alpha = 0.5)

plt.subplot(1,3,2)
plt.plot(x,z, color = 'green', linestyle='dotted', linewidth = 2, marker = 'o', markersize = 10, markerfacecolor='yellow', label = 'Volums mapping')
plt.title('line plot', fontweight = 'bold', color = 'red')
plt.xlabel('X - label')
plt.ylabel('Y - label')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,1))
plt.grid(True, linestyle = '--', alpha = 0.5)

plt.subplot(1,3,3)
plt.plot(x,a, color = 'black', linestyle=':', linewidth = 2, marker = 'D', markersize = 10, markerfacecolor='blue', label = 'exp mapping')
plt.title('line plot', fontweight = 'bold', color = 'red')
plt.xlabel('X - label')
plt.ylabel('Y - label')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,1))
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.tight_layout(pad = 0.1)

plt.show()
```

```python
y1 = np.log(x)
y2 = np.sin(x)
y3 = (np.exp(x) + np.exp(-x))/2
y4 = np.pi*x**3
y5 = np.tan(x)
y6 = np.arccos(x)
y7 = np.tan(x)**2 + np.arcsin(x)
y8 = 3*np.pi*x**3*np.log(x)/np.exp(x)
```

```python
plt.figure(figsize = (15,15))

plt.subplot(2,4,1)
plt.plot(x,y1, color = 'red', linestyle='--', linewidth = 2, marker = '*', markersize = 10, markerfacecolor='black', label = 'np.log(x)')
plt.title('line plot', fontweight = 'bold', color = 'red')
plt.xlabel('X - label')
plt.ylabel('Y - label')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,1))
plt.grid(True, linestyle = '--', alpha = 0.5)

plt.subplot(2,4,2)
plt.plot(x,y2, color = 'green', linestyle='dotted', linewidth = 2, marker = 'o', markersize = 10, markerfacecolor='yellow', label = 'np.sin(x)')
plt.title('line plot', fontweight = 'bold', color = 'red')
plt.xlabel('X - label')
plt.ylabel('Y - label')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,1))
plt.grid(True, linestyle = '--', alpha = 0.5)

plt.subplot(2,4,3)
plt.plot(x,y3, color = 'black', linestyle=':', linewidth = 2, marker = 'D', markersize = 10, markerfacecolor='blue', label = '(np.exp(x) + np.exp(-x))/2')
plt.title('line plot', fontweight = 'bold', color = 'red')
plt.xlabel('X - label')
plt.ylabel('Y - label')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,1))
plt.grid(True, linestyle = '--', alpha = 0.5)

plt.subplot(2,4,4)
plt.plot(x,y4, color = 'blue', linestyle=':', linewidth = 2, marker = 'D', markersize = 10, markerfacecolor='red', label = 'np.pi*x**3')
plt.title('line plot', fontweight = 'bold', color = 'red')
plt.xlabel('X - label')
plt.ylabel('Y - label')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,1))
plt.grid(True, linestyle = '--', alpha = 0.5)

plt.subplot(2,4,5)
plt.plot(x,y5, color = 'gray', linestyle=':', linewidth = 2, marker = 'D', markersize = 10, markerfacecolor='orange', label = 'np.tan(x)')
plt.title('line plot', fontweight = 'bold', color = 'red')
plt.xlabel('X - label')
plt.ylabel('Y - label')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,1))
plt.grid(True, linestyle = '--', alpha = 0.5)

plt.subplot(2,4,6)
plt.plot(x,y6, color = 'orange', linestyle=':', linewidth = 2, marker = 'D', markersize = 10, markerfacecolor='purple', label = 'np.arccos(x)')
plt.title('line plot', fontweight = 'bold', color = 'red')
plt.xlabel('X - label')
plt.ylabel('Y - label')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,1))
plt.grid(True, linestyle = '--', alpha = 0.5)

plt.subplot(2,4,7)
plt.plot(x,y7, color = 'purple', linestyle=':', linewidth = 2, marker = 'D', markersize = 10, markerfacecolor='brown', label = 'np.tan(x)**2 + np.arcsin(x)')
plt.title('line plot', fontweight = 'bold', color = 'red')
plt.xlabel('X - label')
plt.ylabel('Y - label')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,1))
plt.grid(True, linestyle = '--', alpha = 0.5)

plt.subplot(2,4,8)
plt.plot(x,y8, color = 'cyan', linestyle=':', linewidth = 2, marker = 'D', markersize = 10, markerfacecolor='orange', label = '3*np.pi*x**3*np.log(x)/np.exp(x)')
plt.title('line plot', fontweight = 'bold', color = 'red')
plt.xlabel('X - label')
plt.ylabel('Y - label')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,1))
plt.grid(True, linestyle = '--', alpha = 0.5)

plt.tight_layout(pad = 0.2)
plt.show()
```
