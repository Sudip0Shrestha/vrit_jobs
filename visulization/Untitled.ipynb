```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fruits = ['Apple','Banana','Mango','Orange','Grapes']
sales = [30,25,20,15,10]

sns.set_style('darkgrid')

explode = [0.100]*len(fruits)
colors = sns.color_palette('Set2', len(fruits))
plt.figure(figsize = (10,10), facecolor = 'orange')

wedges, texts, autotexts = plt.pie(
    sales,
    labels=fruits,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    explode=explode,
    shadow=True,
    wedgeprops={'edgecolor':'black'},
    textprops={'fontsize':15, 'weight':'bold'}
)

plt.text(0,0,'Fruits Distribution',ha='center',va='center',fontsize=18,weight='bold')

for autotext in autotexts:
    autotext.set_color('black')

plt.title("Fruit Sales Distribution", fontsize = 25, weight = 'bold')

plt.legend(loc='upper right', bbox_to_anchor=(1.3,1))

plt.show()
```

```python
plt.figure(figsize = (4,4))
plt.pie(
    sales,
    labels=fruits,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    radius = 2
)
plt.pie([422], colors = 'w', radius = 1)
plt.show()
```

```python
data = [10,12,15,18,20,22,24,30,35,50]
plt.figure(figsize = (5,5))
sns.boxplot(data=data)

plt.title("Boxplot Example")
plt.show()
```

```python
data = [10, 12, 15, 18, 20, 22, 24, 30, 35, 50]

sns.violinplot(y=data)

plt.title("Violin Plot Example")
plt.show()
```

```python
data = [10,12,15,18,20,22,24,30,35,50]

sns.histplot(data, kde=True)

plt.title("Histogram with KDE")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
