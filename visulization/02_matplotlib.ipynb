```python
name = ['Ram', 'Shyam', 'Hari', 'Sita', 'Gita', 'Rita','you']
marks = [78, 85, 69, 92, 88, 74, 84]
```

```python
bars = (plt.bar(
    name,
    marks,
    edgecolor='black',
    color=['red', 'green', 'orange', 'yellow', 'blue', 'pink']
))
plt.title('Bat Plot', fontsize = 16, fontweight = 'bold')
plt.xlabel('Name of the students')
plt.ylabel('Marks of the students')
plt.xticks(rotation = 90)
plt.grid(axis='y', linestyle='--', alpha=0.9)

for bar in bars:
    height = bar.get_height()
    print('x',bar.get_x())
    print('-'*60)
    print('width',bar.get_width())
    print('-'*60)
    print('height',height)
    print('-'*60)
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f'{height}',
        ha = 'center',
        va = 'bottom',
        fontsize = 10,
        fontweight = 400
        
    )
        
plt.tight_layout()  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,0.1,0.5,1])

y1 = np.log(x+1) + 6*np.pi*x**3
y2 = np.arcsin(x) + np.arccos(x)
y3 = np.sin(x) + 2*x + np.pi*x
y4 = (1/np.cos(x))**2 + (3*x - np.exp(-x))
y5 = np.sin(x)
y6 = np.exp(x)

# reflection
y7  = -y1
y8  = -y2
y9  = -y3
y10 = -y5
y11 = -y6

ys = [y1,y2,y3,y4,y5,y6,y7,y8,y9,y10,y11]

fig, axes = plt.subplots(7,7, figsize=(14,10))

# turn off all axes
for ax in axes.flat:
    ax.axis("off")

positions = [
    (0,0),   #1
    (1,1),   #2
    (2,2),   #3
    (3,3),   #4
    (4,4),   #5
    (5,5),   #6
    (6,0),   #7
    (5,1),   #8
    (4,2),   #9
    (2,4),   #10
    (1,5)    #11
]

for i,(r,c) in enumerate(positions):
    axes[r,c].axis("on")
    axes[r,c].plot(x, ys[i], marker='o', label=f'y{i+1}')
    axes[r,c].set_title(f'Plot {i+1}')
    axes[r,c].legend()

plt.tight_layout()
plt.show()
```
