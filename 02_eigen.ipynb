```python
import numpy as np
import math
```

```python
a = np.array([[2, 3, 1],
              [4, 1, 5],
              [6, 2, 3]])

b = np.array([[1, 4, 2],
              [3, 2, 6],
              [5, 1, 3]])
c = np.array([[3, 1, 4],
              [2, 5, 1],
              [6, 3, 2]])
ai = (np.linalg.inv(a))
bi = (np.linalg.inv(b))
ci = (np.linalg.inv(c))
at = (np.einsum('ij->ji',a))
bt = (np.einsum('ij->ji',b))
ct = (np.einsum('ij->ji',c))
ta = np.einsum('ii',a)
tb = np.einsum('ii',b)
tc = np.einsum('ii',c)
```

```python
result1 = np.einsum('ij, jk -> ik', at,b)
```

```python
result1
```

```python
result2 = (np.einsum('ij, jk -> ik',bi, ct)) + (np.einsum('ij, jk ->ik', at, c))
```

```python
result2
```

```python
result3 = ((np.einsum('ij, jk -> ik', a, ct)) - (np.einsum('ij, jk->ik', (np.einsum('ij, ij -> ij', at,b)),ci)))
```

```python
result3
```

```python
result4 = (np.pi * a) - np.linalg.det(c)*bt
```

```python
result4
```

```python
result5 = (a + bt)@(c-at)
```

```python
result5
```

```python
result6 = (bi*ct + (at@c)*b)
```

```python
result7 = (np.linalg.inv(a+c) @ bt - at @bi) - ((b*ct)@ai + 5*ta*c)
```

```python
result7
```

```python
result8 = np.linalg.det(a)*(bt@c) - 3*np.linalg.inv(b+c)@at + np.pi**2*(at +bt + ct)@(2*a - 3*b + c)
```

```python
result8
```

```python
result9 = 7*(a@b@c) - np.pi**3 * bt @ ai @ ct + np.linalg.det(b)*np.linalg.inv(a+c) @(at -ct)
```

```python
result10 = 4*tc*(a@b - b@c) + np.linalg.inv(at + bt) @ (2*c - np.pi*a)
```

```python
result11 = -6 * np.linalg.det(a+b)*ct + (a*b*c)@ np.linalg.inv(a+b+c)
```
