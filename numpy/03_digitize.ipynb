Find the SVD of given matrix $$
A =
\begin{bmatrix}
1 & 2 & 7 & 9 \\
8 & 10 & 12 & 5 \\
6 & 8 & 9 & 10
\end{bmatrix}
$$

```python
A = np.array([[1, 2, 7, 9],
              [8,10,12,5],
              [6, 8, 9,10]])
```

```python
AT = np.einsum('ij->ji', A)
ATA = np.einsum('ij,jk -> ik', AT, A)
eigenvalues, V = np.linalg.eig(ATA)
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
V = V[:, idx]
singular_values = np.sqrt(eigenvalues)
Sigma = np.zeros((3,4))

for i in range(3):
    Sigma[i,i] = singular_values[i]

U = np.zeros((3,3))

for i in range(3):
    Av = np.einsum('ij,j->i', A, V[:,i])
    U[:,i] = Av / singular_values[i]
    VT = np.einsum('ij->ji', V)

US = np.einsum('ij,jk->ik', U, Sigma)
A_reconstructed = np.einsum('ij,jk->ik', US, VT)

print("Reconstructed A =\n", A_reconstructed)
```
