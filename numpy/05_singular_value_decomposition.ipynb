$$
A =
\begin{bmatrix}
3 & 1 \\
1 & 3 \\
\end{bmatrix}
$$

```python
import numpy as np
```

```python
A = np.array([[3, 1],
              [1, 3]])
```

```python
AtA = A.T @ A
AtA
```

```python
eigenvalue, eigenvector = np.linalg.eig(AtA)
eigenvalue, eigenvector
```

```python
singular_values = np.sqrt(np.sort(eigenvalue)[::-1])
singular_value = np.diag(singular_values)
```

```python
singular_value
```

```python
U = (A @ eigenvector[:,0])/singular_values[:1]
```

```python
U
```
