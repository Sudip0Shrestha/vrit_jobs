```python
import numpy as np
import sympy as sp
from scipy import integrate
```

```python
x = sp.Symbol('x')
f1 = x**6 + 2*x**3 + 4*x**8 + 894

f1_prime = sp.diff(f1,x)
print("f'(x) = ", f1_prime)
```

```python
f2 = sp.sin(x) + sp.cos(x) + sp.tan(x)
f2_prime = sp.diff(f2,x)

f2_prime
```

```python
f3 = sp.exp(x)
f4 = sp.log(x)
a = sp.Symbol('a', postive = True)
f5 = a**x
print("f'3 = ", sp.diff(f3,x))
print("f'4 = ", sp.diff(f4,x))
print("f'5 = ", sp.diff(f5,x))
```

```python
f_chain_rule = sp.sin(x**3 + 3*x)
print("f_chain_rule: ", sp.diff(f_chain_rule, x))
```

```python
f_prod = x**2*sp.exp(x)
y = sp.diff(f_prod,x)
sp.factor(y)
```

```python
f_que = sp.sin(x)/x
sp.diff(f_que,x)
```

```python
f = sp.log(x,10)
sp.diff(f,x)
```

```python
f1 =  32*x**7 + 6*x**5 + 6*x**2
print(f"f1: {sp.integrate(f1,x)} + C")
```

```python
y = sp.Symbol('y')
a = sp.Symbol('a')
f = a**y
print(sp.integrate(f,y),"+C")
```

```python
print(sp.integrate(1/x,x), "+c")
```

```python
f_parts = x**2*sp.exp(x)
integral_parts = sp.integrate(f_parts,x)
print(integral_parts, "+c")
print(sp.factor(integral_parts), "+C")
```

```python
print(sp.simplify(sp.diff(integral_parts,x)))
```

```python
inte = x**2
sp.integrate(inte, (x,0,1))
```

```python
I = 8*sp.cos(2*x)/(((sp.sin(x)**2 + sp.cos(x)**2)/(sp.sin(x)*sp.cos(x)))**3)
simp = sp.simplify(I)
fun = sp.lambdify(x, simp, 'numpy')

result, error = integrate.quad(fun, np.pi/12, np.pi/4)
print("Numerical Result :", result)
print("Entimated error :", error)

# integral_parts = (sp.integrate(I,(x, sp.pi/12, sp.pi/4)))
# print(sp.factor(integral_parts), "+ C")
```
