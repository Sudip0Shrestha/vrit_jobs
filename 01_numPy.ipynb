# create a numpy array for following data from 10 to 1000 andcalculate covarience and compare the value and result generated from scratch and numpy

```python
x = np.random.randint(10,1000,20)
y = np.random.randint(10,1000,20)
```

```python
x
```

```python
y
```

## Calculating Covarience from scratch

```python
sum_x = 0
sum_y = 0
cov = 0
sum_cov = 0
n = len(x)
for i in range(n):
    sum_x += x[i]
    sum_y += x[i]
mean_x = sum_x/n
mean_y = sum_y/n

for i in range(n):
    sum_cov += (x[i] - mean_x)*(y[i] - mean_y)
cov = sum_cov/(n)
cov
    
```

### Calculating covarience using numpy

```python
cov_np = np.cov(x,y)[0,1]
cov_np
```

```python
print(f"Covarience with manually: {cov}")
print(f"Covarience with numpy: {cov_np}")
print(f"the diff on manually and using numpy is {abs(cov-cov_np)}")
```

#### The difference occurs because the manual calculation divides by n (population covariance), while NumPy by default divides by (n-1), which calculates the sample covariance.
